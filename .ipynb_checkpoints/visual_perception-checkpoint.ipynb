## 1 - Drivable Space Estimation Using Semantic Segmentation Output

Now that we know the prerequisites for this notebook, let us hit the road! We start with the implementation of drivable space estimation. We know the following data:
- Semantic segmentation output
- Camera caliberation matrix $K$
- Depth per pixel

### 1.1 - Estimating the x, y, and z coordinates of every pixel in the image:
Equations to get the required 3D coordinates are:

$$z = depth$$

$$x = \frac{(u - c_u) * z}{f} \tag{1}$$

$$y = \frac{(v - c_v) * z}{f} \tag{2}$$

Here, $c_u$, $c_v$, and $f$ are the intrinsic caliberation parameters found in the camera caliberation matrix $K$ such that:

$$K = \begin{pmatrix} f & 0 & c_u \\ 0 & f & c_v \\ 0 & 0 & 1 \end{pmatrix}$$

```python
def xy_from_depth(depth, k):
    """
    Computes the x, and y coordinates of every pixel in the image using the depth map and the calibration matrix.

    Arguments:
    depth -- tensor of dimension (H, W), contains a depth value (in meters) for every pixel in the image.
    k -- tensor of dimension (3x3), the intrinsic camera matrix

    Returns:
    x -- tensor of dimension (H, W) containing the x coordinates of every pixel in the camera coordinate frame.
    y -- tensor of dimension (H, W) containing the y coordinates of every pixel in the camera coordinate frame.
    """
    
    # Get shape of the depth tensor
    depth_shape = depth.shape
    
    # Grab required parameters from the K matrix
    fx = k[0,0]
    fy = k[1,1]
    cu = k[0,2]
    cv = k[1,2]
    
    # Generalize the grid coordinates to the shape of the depth map
    x = np.zeros_like(depth)
    y = np.zeros_like(depth)
    
    # Compute x and y coordinates
    for i in range(depth.shape[0]):
        for j in range(depth.shape[1]):
            x[i,j] = ((j - cu) * depth[i,j]) / fx
            y[i,j] = ((i - cv) * depth[i,j]) / fy
            
    return x, y
```

## 1.2 - Estimating the Ground Plane Using RANSAC:
In the context of self-driving cars, drivable space includes any space that the car is physically capable of traversing in 3D. The task of estimating the drivable space is equivalent to estimating pixels belonging to the ground plane in the scene. In this section, we will be using RANSAC to estimate the ground plane in the 3D camera coordinate frame from the x, y, and z coordinates estimated above.

The first step is to process the semantic segmentation output to extract the relevant pixel belonging to the class you want to consider as ground. Hence, that class is road class with a mapping index of 7. To extract the x, y, z coordinates of the road, run the following cell:

```python
# Get road mask by choosing pixels in segmentation output with value 7
road_mask = np.zeros(segmentation.shape)
road_mask[segmentation == 7] = 1

# Show road mask
plt.imshow(road_mask)

# Get x, y, z coordinates of the pixels in road mask
x_ground = x[road_mask == 1]
y_ground = y[road_mask == 1]
z_ground = dataset_handler.depth[road_mask == 1]
xyz_ground = np.stack((x_ground, y_ground, z_ground))
```

The next step is to use the extracted x, y, z coordinates of pixels belonging to the road to estimate the ground plane. RANSAC will be used for robustness against outliers.

RANSAC algorithm follows the following steps:
1. Choose a minimum of 3 points from `xyz_ground` at random.
2. Compute the ground plane model using the chosen random points, and the provided function `compute_plane`.
3. Compute the distance from the ground plane model to every point in `xyz_ground`, and compute the number of inliers based on a distance threshold.
4. Check if the current number of inliers is greater than all previous iterations and keep the inlier set with the largest number of points.  
5. Repeat until number of iterations $\geq$ a preset number of iterations, or number of inliers $\geq$ minimum number of inliers.
6. Recompute and return a plane model using all inliers in the final inlier set.

```python
def ransac_plane_fit(xyz_data):
    """
    Computes plane coefficients a,b,c,d of the plane in the form ax+by+cz+d = 0
    using ransac for outlier rejection.

    Arguments:
    xyz_data -- tensor of dimension (3, N), contains all data points from which random sampling will proceed.
    distance_threshold -- Distance threshold from plane for a point to be considered an inlier.

    Returns:
    p -- tensor of dimension (1, 4) containing the plane parameters a,b,c,d
    """
    
    # Set thresholds
    num_iter = 1000                 # RANSAC maximum number of iterations
    min_num_inliers = 45000         # RANSAC minimum number of inliers
    distance_threshold = 0.1        # Maximum distance from point to plane for point to be considered inlier
    
    data_size = xyz_data.shape[1]
    
    x = xyz_data[0,:]
    y = xyz_data[1,:]
    z = xyz_data[2,:]
    
    max_inliers = 0    
    for i in range(num_iter):
        # Choose a minimum of 3 points from xyz_data at random
        inds = np.random.choice(data_size, 15, replace = False)
        
        # Compute plane model
        plane = compute_plane(xyz_data[:, inds])
        
        # Find number of inliers
        dist = np.abs(dist_to_plane(plane, x, y, z))
        num_inliers = np.sum(dist < distance_threshold)
        
        # Check if the current number of inliers is greater than all previous iterations and keep the inlier set with the largest number of points.
        if max_inliers < num_inliers:
            max_inliers = num_inliers
            inlier_set = xyz_data[:, dist < distance_threshold]
            
        # Check if stopping criterion is satisfied and break
        if num_inliers > min_num_inliers:
            break
    
    # Recompute the model parameters using largest inlier set.
    output_plane = compute_plane(inlier_set)
    
    return output_plane
```

```python
p_final = ransac_plane_fit(xyz_ground)
print('Ground Plane: ' + str(p_final))
```

## 2 - Lane Estimation using the Semantic Segmentation Output
Next step is to use the output of semantic segmentation to estimate the lane boundaries of the current lane the self-driving car is using. The task can be divided into two subtasks - lane line estimation, and post-processing through horizontal lane filtering and similar line merging.

### 2.1 - Estimating Lane Boundary Proposals
The first step to perform this task is to estimate any line that qualifies as a lane boundary using the output from semantic segmentation. We call these lines 'proposals'.

```python
def estimate_lane_lines(segmentation_output):
    """
    Estimates lines belonging to lane boundaries. Multiple lines could correspond to a single lane.

    Arguments:
    segmentation_output -- tensor of dimension (H,W), containing semantic segmentation neural network output
    minLineLength -- Scalar, the minimum line length
    maxLineGap -- Scalar, dimension (Nx1), containing the z coordinates of the points

    Returns:
    lines -- tensor of dimension (N, 4) containing lines in the form of [x_1, y_1, x_2, y_2], where [x_1,y_1] and [x_2,y_2] are
    the coordinates of two points on the line in the (u,v) image coordinate frame.
    """
    # Create an image with pixels belonging to lane boundary categories from the output of semantic segmentation
    seg_new = np.zeros(segmentation_output.shape)
    seg_new[segmentation_output == 6] = 255
    seg_new[segmentation_output == 8] = 255
    
    # Perform Edge Detection
    seg_new = cv2.GaussianBlur(seg_new, (5,5), 1)
    edges = cv2.Canny(seg_new.astype(np.uint8), 100, 110, apertureSize = 3)
    
    # Perform line estimation using cv2.HoughLinesP()
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold = 100, minLineLength = 100, maxLineGap = 25)
    lines = np.squeeze(lines,axis=1)
    
    return lines
```

```python
lane_lines = estimate_lane_lines(segmentation)

plt.imshow(dataset_handler.vis_lanes(lane_lines))
```

### 2.2 - Merginfgand Filtering Lane Lines:
The next subtask to perform the estimation of the current lane boundary is to merge redundant lines, and filter out any horizontal lines apparent in the image. Merging redundant lines can be solved through grouping with similar slope and intercept. Horizontal lines can be filtered out through slope thresholding.

```python
def merge_lane_lines(lines):
    """
    Merges lane lines to output a single line per lane, using the slope and intercept as similarity measures.
    Also, filters horizontal lane lines based on a minimum slope threshold.

    Arguments:
    lines -- tensor of dimension (N, 4) containing lines in the form of [x_1, y_1, x_2, y_2],
    the coordinates of two points on the line.

    Returns:
    merged_lines -- tensor of dimension (N, 4) containing lines in the form of [x_1, y_1, x_2, y_2],
    the coordinates of two points on the line.
    """
    # Define thresholds
    slope_similarity_threshold = 0.1
    intercept_similarity_threshold = 40
    min_slope_threshold = 0.3
    
    # Get slope and intercept of lines
    slopes, intercepts = get_slope_intercept(lines)
    
    clusters = []
    current_inds = []
    itr = 0
    
    # Determine lines with slope less than horizontal slope threshold.
    slope_horizontal_filter = np.abs(slopes) > min_slope_threshold
    
    # Iterate over all remaining slopes and intercepts and cluster lines that are close to each other using a slope and intercept threshold.
    for slope, intercept in zip(slopes, intercepts):
        exists_in_cluster = np.array([itr in current for current in current_inds])
        if not exists_in_cluster.any():
            slope_cluster = np.logical_and(slopes < (slope + slope_similarity_threshold), 
                                           slopes > (slope - slope_similarity_threshold))
            
            intercept_cluster = np.logical_and(intercepts < (intercept + intercept_similarity_threshold), 
                                               intercepts > (intercept - intercept_similarity_threshold))
            
            inds = np.argwhere(slope_cluster & intercept_cluster & slope_horizontal_filter).T
            if inds.size:
                current_inds.append(inds.flatten())
                clusters.append(lines[inds])
        itr += 1
        
    # Merge all lines in clusters using mean averaging
    merged_lines = [np.mean(cluster, axis = 1) for cluster in clusters]
    merged_lines = np.squeeze(np.array(merged_lines), axis = 1)
    
    return merged_lines
```

## 3 - Computing minimum distance to impact using the output of 2D Object Detection
The final task is to use 2D object detection output to determine the minimum distance to imapct with objects in the scene. However, the task is complicated by the fact that the 2D detection outputs can be of a high recall, low precision. So to solve this problem we will first use semantic segmentation output to determine which bounding boxes are valid. Then, we will compute the minimum distance to impact using the remaining bounding boxes and the depth image.

```python
detections = dataset_handler.object_detection

plt.imshow(dataset_handler.vis_object_detection(detections))
```

Detections have the format `[category, x_min, y_min, x_max, y_max]`. The Category is a string signifying the classification of the bounding box such as `'Car'`, `'Pedestrian'` or `'Cyclist'`. `[x_min, y_min]` are the coordinates of the top left corner, and `[x_max, y_max]` are the coordinates of the bottom right corners of the object. The score signifies the output of the softmax from the neural network.

```python
print(detections)
```

### 3.1 - Filtering out unreliable detections
First thing you can notice is that a wrong detection occurs on the right side of the image. The interesting observation is that this wrong detection has a high output score of 0.76 for being a car. Furthermore, two bounding boxes are assigned to the vehicle to the left of the image, both with a very high score, greater than 0.9. This behaviour is expected from a high precision, low recall object detector. To solve this problem, the output of the semantic segmentation network has to be used to eliminate unreliable detections.

```python
def filter_detections_by_segmentation(detections, segmentation_output):
    """
    Filter 2D detection output based on a semantic segmentation map.

    Arguments:
    detections -- tensor of dimension (N, 5) containing detections in the form of [Class, x_min, y_min, x_max, y_max, score].
    
    segmentation_output -- tensor of dimension (HxW) containing pixel category labels.
    
    Returns:
    filtered_detections -- tensor of dimension (N, 5) containing detections in the form of [Class, x_min, y_min, x_max, y_max, score].

    """
    # Set ratio threshold
    ratio_threshold = 0.3 # If 1/3 of the total pixels belong to the target category, the detection is correct.
    
    filtered_detections = []
    
    # Main loop over detections
    for detection in detections:
        bounding_box = np.asfarray(detection[1:5])
        # Compute number of pixels belonging to the category for every detection.
        segmentation_crop = segmentation_output[int(bounding_box[1]):int(bounding_box[3]), 
                                                int(bounding_box[0]):int(bounding_box[2])]
        
        category = detection[0]
        height = bounding_box[3] - bounding_box[1]
        width = bounding_box[2] - bounding_box[0]
        area = height * width
                    
        # Divide the computed number of pixels by the area of the bounding box (total number of pixels).
        if category == 'Car'or category == 'Cyclist':
            num_pixels = np.sum(segmentation_crop == 10)
            ratio = num_pixels / area
        elif category == 'Pedestrian':
            num_pixels = np.sum(segmentation_crop == 4)
            ratio = num_pixels / area
        
        # If the ratio is greater than a threshold keep the detection. Else, remove the detection from the list of detections.
        if ratio > ratio_threshold:
            filtered_detections.append(detection)
    filtered_detections = np.asarray(filtered_detections)
            
    return filtered_detections                
```

```python
filtered_detections = filter_detections_by_segmentation(detections, segmentation)
plt.imshow(dataset_handler.vis_object_detection(filtered_detections))
```

### 3.2 - Estimating minimum distance to impact
Now we are ready to calculate the minimum distance to every bounding box in the input detections. This can be performed by simply taking the minimum distance from the pixels in the bounding box to the camera center.

```python
def find_min_distance_to_detection(detections, x, y, z):
    """
    Filter 2D detection output based on a semantic segmentation map.

    Arguments:
    detections -- tensor of dimension (N, 5) containing detections in the form of [Class, x_min, y_min, x_max, y_max, score].
    
    x -- tensor of dimension (H, W) containing the x coordinates of every pixel in the camera coordinate frame.
    y -- tensor of dimension (H, W) containing the y coordinates of every pixel in the camera coordinate frame.
    z -- tensor of dimensions (H,W) containing the z coordinates of every pixel in the camera coordinate frame.
    Returns:
    min_distances -- tensor of dimension (N, 1) containing distance to impact with every object in the scene.

    """
    min_distances = []
    # Compute distance to camera
    center_cam_distance = np.sqrt(x**2 + y**2 + z**2)
    
    for detection in detections:
        bounding_box = np.asfarray(detection[1:5])
        # Compute distance of every pixel in the detection bounds
        center_cam_distance_crop = center_cam_distance[int(bounding_box[1]):int(bounding_box[3]), 
                                                       int(bounding_box[0]):int(bounding_box[2])]
        # Find minimum distance
        min_distances.append(np.min(center_cam_distance_crop))
    min_distances = np.array(min_distances)
    return min_distances
```

```python
min_distances = find_min_distance_to_detection(filtered_detections, x, y, z)
print('Minimum distance to impact is:' + str(min_distances[0]))
```

Run the following cell to visualize the estimated distance along with the 2D detection output.
